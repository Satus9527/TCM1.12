```python
# ä¼˜åŒ–ä»²æ™¯ä¸­åŒ»å’¨è¯¢ç³»ç»Ÿ
class OptimizedZhongjingTCMConsultation:
    def __init__(self, tokenizer, model):
        self.tokenizer = tokenizer
        self.model = model
        self.device = model.device  # è·å–æ¨¡å‹æ‰€åœ¨çš„è®¾å¤‡

        # ä¼˜åŒ–çš„ä¸­åŒ»ä¸“ç”¨promptæ¨¡æ¿
        self.prompt_templates = {
            "diagnosis": "æ ¹æ®ã€Šä¼¤å¯’è®ºã€‹å’Œã€Šé‡‘åŒ®è¦ç•¥ã€‹çš„è¾¨è¯åŸåˆ™ï¼Œåˆ†æä»¥ä¸‹ç—‡çŠ¶ï¼š{}ã€‚è¯·ç»™å‡ºä¸­åŒ»è¯Šæ–­ã€è¯å€™åˆ†æå’Œæ²»ç–—æ–¹å‰‚ï¼š",
            "treatment": "é’ˆå¯¹ä»¥ä¸‹ç—‡çŠ¶ï¼š{}ï¼Œè¯·åŸºäºå¼ ä»²æ™¯çš„ç»æ–¹ç†è®ºï¼Œæ¨èåˆé€‚çš„æ–¹å‰‚å¹¶è¯´æ˜é…ä¼åŸç†ï¼š",
            "syndrome": "æ ¹æ®å…­ç»è¾¨è¯ï¼Œåˆ†æä»¥ä¸‹ç—‡çŠ¶ï¼š{}ã€‚è¯·åˆ¤æ–­å±äºå“ªä¸€ç»ç—…è¯ï¼Œå¹¶ç»™å‡ºç†æ³•æ–¹è¯ï¼š"
        }

    def consult(self, symptoms, consult_type="diagnosis", max_length=120, temperature=0.7):
        """ä¼˜åŒ–çš„ä¸­åŒ»å’¨è¯¢å‡½æ•°"""
        try:
            # ä½¿ç”¨ä¼˜åŒ–çš„promptæ¨¡æ¿
            prompt_template = self.prompt_templates.get(consult_type, self.prompt_templates["diagnosis"])
            prompt = prompt_template.format(symptoms)

            # ç¼–ç è¾“å…¥å¹¶ç§»åŠ¨åˆ°æ¨¡å‹æ‰€åœ¨è®¾å¤‡
            inputs = self.tokenizer(prompt, return_tensors="pt")
            inputs = {key: value.to(self.device) for key, value in inputs.items()}  # å…³é”®ä¿®å¤ï¼

            # ä½¿ç”¨æ›´ä¸¥æ ¼çš„ç”Ÿæˆå‚æ•°
            with torch.no_grad():
                outputs = self.model.generate(
                    **inputs,
                    max_new_tokens=max_length,
                    temperature=temperature,
                    do_sample=True,
                    top_p=0.85,  # é™ä½top_pï¼Œå‡å°‘éšæœºæ€§
                    pad_token_id=self.tokenizer.eos_token_id,
                    repetition_penalty=1.2,  # å¢åŠ é‡å¤æƒ©ç½š
                    no_repeat_ngram_size=4,  # é¿å…4-gramé‡å¤
                    # ç§»é™¤ early_stopping å‚æ•°
                )

            # è§£ç ç»“æœ
            response = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

            # æå–æœ‰æ•ˆå›ç­”ï¼ˆç§»é™¤promptéƒ¨åˆ†ï¼‰
            if prompt in response:
                response = response.replace(prompt, "").strip()

            return response

        except Exception as e:
            return f"å’¨è¯¢è¿‡ç¨‹ä¸­å‡ºç°é”™è¯¯: {e}"

    def strict_consult(self, symptoms, max_length=100):
        """æ›´ä¸¥æ ¼çš„ä¸­åŒ»å’¨è¯¢ï¼Œé¿å…ç°ä»£åŒ»å­¦å†…å®¹"""
        try:
            # ä½¿ç”¨æ›´å…·ä½“çš„prompt
            prompt = f"è¯·ä»¥å¼ ä»²æ™¯ã€Šä¼¤å¯’æ‚ç—…è®ºã€‹çš„ç»æ–¹ç†è®ºä¸ºåŸºç¡€ï¼Œåˆ†æä»¥ä¸‹ç—‡çŠ¶å¹¶ç»™å‡ºçº¯ä¸­åŒ»çš„æ²»ç–—å»ºè®®ï¼š{symptoms}ã€‚é¿å…æåŠç°ä»£åŒ»å­¦å†…å®¹ï¼š"

            # ç¼–ç è¾“å…¥å¹¶ç§»åŠ¨åˆ°æ¨¡å‹æ‰€åœ¨è®¾å¤‡
            inputs = self.tokenizer(prompt, return_tensors="pt")
            inputs = {key: value.to(self.device) for key, value in inputs.items()}  # å…³é”®ä¿®å¤ï¼

            with torch.no_grad():
                outputs = self.model.generate(
                    **inputs,
                    max_new_tokens=max_length,
                    temperature=0.6,  # æ›´ä½çš„æ¸©åº¦ï¼Œå‡å°‘éšæœºæ€§
                    do_sample=True,
                    top_p=0.8,
                    pad_token_id=self.tokenizer.eos_token_id,
                    repetition_penalty=1.3,
                    no_repeat_ngram_size=5,
                    # ç§»é™¤ early_stopping å‚æ•°
                )

            response = self.tokenizer.decode(outputs[0], skip_special_tokens=True)

            # è¿‡æ»¤æ‰åŒ…å«ç°ä»£åŒ»å­¦æœ¯è¯­çš„å›ç­”
            modern_terms = ["å¸ƒæ´›èŠ¬", "æŠ—ç”Ÿç´ ", "åŒ»é™¢", "æ£€æŸ¥", "æ‰‹æœ¯", "è¥¿åŒ»", "ç°ä»£åŒ»å­¦", "åŒ–éªŒ"]
            if any(term in response for term in modern_terms):
                return "æ¨¡å‹å›ç­”åŒ…å«ç°ä»£åŒ»å­¦å†…å®¹ï¼Œå»ºè®®é‡æ–°ç”Ÿæˆæˆ–å’¨è¯¢ä¸“ä¸šä¸­åŒ»å¸ˆã€‚"

            return response.replace(prompt, "").strip()

        except Exception as e:
            return f"å’¨è¯¢è¿‡ç¨‹ä¸­å‡ºç°é”™è¯¯: {e}"

# åˆ›å»ºä¼˜åŒ–çš„å’¨è¯¢ç³»ç»Ÿ
print("ğŸ”§ åˆ›å»ºä¼˜åŒ–çš„ä»²æ™¯ä¸­åŒ»å’¨è¯¢ç³»ç»Ÿ...")
optimized_system = OptimizedZhongjingTCMConsultation(tokenizer, model)

# æµ‹è¯•ä¼˜åŒ–åçš„ç³»ç»Ÿ
print("\nğŸ§ª ä¼˜åŒ–ç³»ç»Ÿæµ‹è¯•:")

test_cases = [
    "å¤´ç—›å‘çƒ­ï¼Œæ¶å¯’æ— æ±—",
    "å’³å—½ç—°å¤šï¼Œè‰²ç™½æ¸…ç¨€",
    "å‘çƒ­æ¶å¯’ï¼Œèº«ä½“ç–¼ç—›"
]

for i, symptoms in enumerate(test_cases, 1):
    print(f"\næ¡ˆä¾‹ {i}: {symptoms}")

    # ä½¿ç”¨ä¸¥æ ¼å’¨è¯¢æ¨¡å¼
    advice = optimized_system.strict_consult(symptoms)
    print(f"ä¼˜åŒ–å»ºè®®: {advice}")

    print("-" * 80)
```

```python
import json
import pandas as pd
from datasets import Dataset

# åŠ è½½æ‚¨çš„æ•°æ®é›†
with open('herbs.json', 'r', encoding='utf-8') as f:
    herbs_data = json.load(f)

with open('prescriptions.json', 'r', encoding='utf-8') as f:
    prescriptions_data = json.load(f)

print(f"è¯ææ•°é‡: {len(herbs_data)}")
print(f"è¯æ–¹æ•°é‡: {len(prescriptions_data)}")

# åˆ›å»ºè®­ç»ƒæ•°æ®
def create_training_samples(herbs, prescriptions):
    samples = []

    # 1. è¯æçŸ¥è¯†è®­ç»ƒæ ·æœ¬
    for herb in herbs:
        prompt = f"è¯·è¯¦ç»†ä»‹ç»ä¸­è¯{herb['name']}ï¼š"
        response = f"{herb['name']}å±äº{herb['category']}ï¼Œæ€§{herb['property']}ï¼Œå‘³{'ã€'.join(herb['flavor'])}ï¼Œå½’{'ã€'.join(herb['meridian'])}ã€‚åŠŸæ•ˆï¼š{herb['efficacy']}ã€‚å¸¸ç”¨å‰‚é‡ï¼š{herb['dosage'][0]}-{herb['dosage'][1]}gã€‚"
        if 'contraindications' in herb:
            response += f"ç¦å¿Œï¼š{'ã€'.join(herb['contraindications'])}ã€‚"
        samples.append({"prompt": prompt, "response": response})

    # 2. è¯æ–¹çŸ¥è¯†è®­ç»ƒæ ·æœ¬
    for prescription in prescriptions:
        # è¯æ–¹åŸºæœ¬ä¿¡æ¯
        prompt = f"è¯·ä»‹ç»{prescription['æ–¹å']}ï¼š"
        response = f"{prescription['æ–¹å']}å‡ºè‡ª{prescription['å‡ºå¤„']}ï¼ŒåŠŸæ•ˆï¼š{prescription['åŠŸæ•ˆ']}ã€‚ä¸»æ²»ï¼š{prescription['ä¸»æ²»']}ã€‚"

        # ç»„æˆå’Œé…ä¼
        response += "ç»„æˆï¼š"
        for component in prescription['ç»„æˆ']:
            response += f"{component['è¯æ']}{component['å‰‚é‡']}ï¼ˆ{component['è§’è‰²']}è¯ï¼‰ã€"
        response = response.rstrip('ã€') + "ã€‚"

        # é…ä¼ç‰¹ç‚¹
        response += f"é…ä¼ç‰¹ç‚¹ï¼š{prescription['é…ä¼ç‰¹ç‚¹']}ã€‚"

        samples.append({"prompt": prompt, "response": response})

        # 3. ç—‡çŠ¶-è¯æ–¹å¯¹åº”æ ·æœ¬
        symptoms_prompt = f"ç—‡çŠ¶ï¼š{prescription['ä¸»æ²»'].split('ã€‚')[0]}ï¼Œè¯·é—®åº”è¯¥ç”¨ä»€ä¹ˆæ–¹å‰‚ï¼Ÿ"
        symptoms_response = f"æ ¹æ®ç—‡çŠ¶åˆ†æï¼Œå»ºè®®ä½¿ç”¨{prescription['æ–¹å']}ã€‚è¯¥æ–¹å‰‚{prescription['åŠŸæ•ˆ']}ï¼Œä¸»æ²»{prescription['ä¸»æ²»'].split('ã€‚')[0]}ã€‚"
        samples.append({"prompt": symptoms_prompt, "response": symptoms_response})

    return samples

# åˆ›å»ºè®­ç»ƒæ•°æ®
training_samples = create_training_samples(herbs_data, prescriptions_data)
print(f"åˆ›å»ºäº† {len(training_samples)} ä¸ªè®­ç»ƒæ ·æœ¬")

# è½¬æ¢ä¸ºDatasetæ ¼å¼
train_dataset = Dataset.from_list(training_samples)

# æŸ¥çœ‹æ ·æœ¬ç¤ºä¾‹
for i in range(6):
    print(f"\næ ·æœ¬ {i+1}:")
    print(f"æç¤º: {train_dataset[i]['prompt']}")
    print(f"å›ç­”: {train_dataset[i]['response']}")
```

```python
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    TrainingArguments,
    Trainer,
    DataCollatorForLanguageModeling
)
from peft import LoraConfig, get_peft_model, TaskType
import torch

# æ¨¡å‹é…ç½®
model_name = "CMLL/ZhongJing-2-1_8b"

# åŠ è½½tokenizer
tokenizer = AutoTokenizer.from_pretrained(
    model_name,
    trust_remote_code=True,
    padding_side="right"
)

if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# æ— é‡åŒ–åŠ è½½ï¼ˆæœ€ç¨³å®šï¼‰
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto",
    trust_remote_code=True,
    # ç§»é™¤æ‰€æœ‰é‡åŒ–ç›¸å…³å‚æ•°
)

print("âœ… æ— é‡åŒ–æ¨¡å‹åŠ è½½æˆåŠŸ!")
print(f"æ¨¡å‹å‚æ•°æ•°é‡: {model.num_parameters():,}")

# é…ç½®LoRA
lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=8,
    lora_alpha=32,
    lora_dropout=0.1,
    target_modules=["q_proj", "v_proj", "k_proj", "o_proj"]
)

# åº”ç”¨LoRA
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

```python
# åˆ›å»ºå®Œæ•´çš„ä¸­åŒ»å’¨è¯¢ç³»ç»Ÿç±»
class ZhongjingTCMExpertSystem:
    def __init__(self, tokenizer, model):
        self.tokenizer = tokenizer
        self.model = model
        self.device = model.device

        # ä¸“ä¸šçš„ä¸­åŒ»å’¨è¯¢æ¨¡æ¿
        self.consultation_templates = {
            "diagnosis": "æ ¹æ®å¼ ä»²æ™¯ã€Šä¼¤å¯’æ‚ç—…è®ºã€‹ç†è®ºï¼Œåˆ†æä»¥ä¸‹ç—‡çŠ¶ï¼š{}ã€‚è¯·ç»™å‡ºï¼š\n1. ä¸­åŒ»è¯Šæ–­\n2. è¯å€™åˆ†æ\n3. æ²»ç–—åŸåˆ™\n4. æ¨èæ–¹å‰‚\n5. æ–¹è§£ï¼š",
            "herb_query": "è¯·è¯¦ç»†ä»‹ç»ä¸­è¯{}çš„åŠŸæ•ˆã€æ€§å‘³ã€å½’ç»ã€ç”¨æ³•ç”¨é‡åŠæ³¨æ„äº‹é¡¹ï¼š",
            "prescription_query": "è¯·è§£ææ–¹å‰‚{}çš„ç»„æˆã€åŠŸæ•ˆã€ä¸»æ²»ã€é…ä¼ç‰¹ç‚¹åŠç°ä»£åº”ç”¨ï¼š",
            "syndrome_differentiation": "æ ¹æ®å…­ç»è¾¨è¯ï¼Œåˆ†æç—‡çŠ¶{}å±äºå“ªä¸€ç»ç—…è¯ï¼Œå¹¶ç»™å‡ºç†æ³•æ–¹è¯ï¼š"
        }

    def consult(self, query, consult_type="diagnosis", max_length=200):
        """ä¸­åŒ»ä¸“å®¶å’¨è¯¢"""
        try:
            if consult_type in self.consultation_templates:
                prompt = self.consultation_templates[consult_type].format(query)
            else:
                prompt = f"è¯·åŸºäºå¼ ä»²æ™¯ç»æ–¹ç†è®ºå›ç­”ï¼š{query}"

            # ç¼–ç è¾“å…¥
            inputs = self.tokenizer(prompt, return_tensors="pt").to(self.device)

            # ç”Ÿæˆå›ç­”
            with torch.no_grad():
                outputs = self.model.generate(
                    **inputs,
                    max_new_tokens=max_length,
                    temperature=0.7,
                    do_sample=True,
                    top_p=0.9,
                    pad_token_id=self.tokenizer.eos_token_id,
                    repetition_penalty=1.2
                )

            # è§£ç ç»“æœ
            response = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
            answer = response.replace(prompt, "").strip()

            return {
                "success": True,
                "question": query,
                "answer": answer,
                "type": consult_type
            }

        except Exception as e:
            return {
                "success": False,
                "error": str(e),
                "question": query
            }

    def batch_consult(self, queries):
        """æ‰¹é‡å’¨è¯¢"""
        results = []
        for query in queries:
            result = self.consult(query)
            results.append(result)
        return results

# åˆ›å»ºä¸“å®¶ç³»ç»Ÿ
print("ğŸ”§ åˆ›å»ºä»²æ™¯ä¸­åŒ»ä¸“å®¶ç³»ç»Ÿ...")
expert_system = ZhongjingTCMExpertSystem(tokenizer, model)

# æµ‹è¯•ä¸“å®¶ç³»ç»Ÿ
print("\nğŸ§ª ä¸“å®¶ç³»ç»Ÿæµ‹è¯•:")
test_queries = [
    ("å¤´ç—›å‘çƒ­ï¼Œæ¶å¯’æ— æ±—ï¼Œå‘¨èº«ç–¼ç—›", "diagnosis"),
    ("äººå‚", "herb_query"),
    ("æ¡‚ææ±¤", "prescription_query"),
    ("å‘çƒ­æ¶å¯’ï¼Œå¤´ç—›é¡¹å¼ºï¼Œè„‰æµ®", "syndrome_differentiation")
]

for symptoms, consult_type in test_queries:
    print(f"\nğŸ“‹ å’¨è¯¢ç±»å‹: {consult_type}")
    print(f"ç—‡çŠ¶: {symptoms}")

    result = expert_system.consult(symptoms, consult_type)

    if result["success"]:
        print(f"ğŸ’¡ ä¸“å®¶å»ºè®®:\n{result['answer']}")
    else:
        print(f"âŒ å’¨è¯¢å¤±è´¥: {result['error']}")

    print("-" * 100)
```
